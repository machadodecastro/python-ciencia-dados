```python
# Importando os pacotes
import seaborn as sea
import matplotlib.pyplot as plt
import numpy as np

# Carregando o conjunto de dados Iris
iris = sea.load_dataset('iris')

# Criando um gráfico de dispersão usando Seaborn
sea.scatterplot(x='sepal_length', y='sepal_width', data=iris, hue='species')

# Adicionando título ao gráfico
plt.title('Gráfico de Dispersão de Comprimento e Largura da Sépala')

# Exibindo o gráfico
plt.show()
```
