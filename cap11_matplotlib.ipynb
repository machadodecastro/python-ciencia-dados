### Gráficos de Barras

Um gráfico de barras é um gráfico com barras retangulares. O comprimento de cada barra é proporcional ao valor que representa. Use gráficos de barras para comparar uma ou duas variáveis usando um único valor.


```python
x = [1,3,5,7,9]
y = [2,4,6,8,10]

plt.bar(x, y, label = 'Barras', color = 'b')
plt.legend()
plt.show()
```

```python
x2 = [1,3,5,7,9]
y2 = [7,8,2,4,2]

plt.bar(x, y, label = 'Barras1', color = 'b')
plt.bar(x2, y2, label = 'Barras2', color = 'y')
plt.legend()
plt.show()
```

### Scatterplot

Os diagramas de dispersão ou gráficos de dispersão são representações de dados de duas ou mais variáveis que são organizadas em um gráfico. O gráfico de dispersão utiliza coordenadas cartesianas para exibir valores de um conjunto de dados.

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,5,6,8,4,8]

plt.scatter(x, y, label = 'Pontos', color = 'r', marker = 'o', s = 100)
plt.legend()
plt.show()
```

### Stack Plots

Stackplots desenham vários conjuntos de dados como áreas empilhadas verticalmente. Isto é útil quando os valores dos dados individuais e, adicionalmente, o seu valor cumulativo são de interesse.

```python
dias = [1,2,3,4,5]
vendas = [7,8,6,77,7]
perdas = [2,3,4,5,3]
lucros = [7,8,7,2,2]
custos = [8,5,7,8,13]

plt.stackplot(dias, vendas, perdas, lucros, custos, colors = ['m','c','r','k','b'])
plt.show()
```

### Pie Chart

Gráfico de setores ou gráfico circular, como é tradicionalmente chamado gráfico de pizza é um diagrama circular em que os valores de cada categoria estatística representada são proporcionais às respectivas medidas dos ângulos. É um gráfico para ser utilizado com poucas variáveis, caso contrário fica muito confuso de ler.

```python
pizzas = [7, 2, 2, 13]
sabores = ['calabresa','portuguesa','atum','toscana']
colunas = ['c','m','r','k']
```

```python
plt.pie(pizzas, labels = sabores, colors = colunas, startangle = 90, shadow = True, explode = (0,0.1,0,0))
plt.show()
```

O módulo pylab faz parte da biblioteca matplotlib em Python. Ele combina matplotlib.pyplot (para criar visualizações) e pacotes numpy (para criar, modificar e manipular conjuntos de dados) para fornecer aos usuários uma maneira conveniente de traçar gráficos.

```python
# Gráficos com 2 figuras
x = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # eixos da figura principal
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # eixos da figura secundária

# Figura principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura 1')

# Figura secundária
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Figura 2');
```

```python
# Gráficos em Paralelo
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título ABC')
    
fig.tight_layout()
```

### Histogramas

Um histograma é uma espécie de gráfico de barras que demonstra uma distribuição de frequências. No histograma, a base de cada uma das barras representa uma classe e a altura representa a quantidade ou frequência absoluta com que o valor de cada classe ocorre.

```python
# Histogramas
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].hist(n)
axes[0].set_title("Histograma Padrão")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title("Histograma Cumulativo")
axes[1].set_xlim((min(n), max(n)));
```

### Mapas de Calor


Mapa de calor ou heat maps tem como objetivo principal coletar os dados de um dataset e entender quais são os pontos mais "quentes" e "frios" em uma escala de cores diferentes. 

```python
# Mapas de Calor
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def ColorMap(phi_m, phi_p):
    return ( + alpha - 2 * np.cos(phi_p)*cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p))

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = ColorMap(X, Y).T

fig, ax = plt.subplots()

p = ax.pcolor(X/(2*np.pi), Y/(2*np.pi), Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p, ax=ax)
```
