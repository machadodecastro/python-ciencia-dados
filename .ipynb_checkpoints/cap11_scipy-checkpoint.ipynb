```python
# Integrando com matplotlib
from pylab import *
%matplotlib inline
```

```python
def dy(y, t, zeta, w0):
    x, p = y[0], y[1]
    
    dx = p
    dp = -2 * zeta * w0 * p - w0**2 * x

    return [dx, dp]


y0 = [1.0, 0.0]

t = linspace(0, 10, 1000)
w0 = 2*pi*1.0


y1 = odeint(dy, y0, t, args=(0.0, w0)) 
y2 = odeint(dy, y0, t, args=(0.2, w0)) 
y3 = odeint(dy, y0, t, args=(1.0, w0)) 
y4 = odeint(dy, y0, t, args=(5.0, w0)) 

fig, ax = subplots()
ax.plot(t, y1[:,0], 'k', label="Não Abafado", linewidth=0.25)
ax.plot(t, y2[:,0], 'r', label="Pouco Abafado")
ax.plot(t, y3[:,0], 'b', label="Criticamente Abafado")
ax.plot(t, y4[:,0], 'g', label="Perigosamente Abafado")
ax.legend();
```

### scipy.optimize.curve_fit()

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
```

```python
# Definindo os dados a serem ajustados com algum ruído
xdata = np.linspace(0, 4, 50)
y = func(xdata, 2.5, 1.3, 0.5)
rng = np.random.default_rng()
y_noise = 0.2 * rng.normal(size=xdata.size)
ydata = y + y_noise
plt.plot(xdata, ydata, 'b-', label='data')
```

```python
# Ajustando os parâmetros a, b, c da função func
opt, pcov = curve_fit(func, xdata, ydata)
popt
np.array([2.56274217, 1.37268521, 0.47427475])
plt.plot(xdata, func(xdata, *popt), 'r-',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
```

```python
# Restringindo a otimização à região de 0 <= a <= 3, 0 <= b <= 1 e 0 <= c <= 0,5
popt, pcov = curve_fit(func, xdata, ydata, bounds=(0, [3., 1., 0.5]))
popt
np.array([2.43736712, 1.        , 0.34463856])
plt.plot(xdata, func(xdata, *popt), 'g--',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

### scipy.signal.convolve()

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

sig = np.repeat([0., 1., 0.], 100)
win = signal.windows.hann(50)
filtered = signal.convolve(sig, win, mode='same') / sum(win)
```

```python
fig, (ax_orig, ax_win, ax_filt) = plt.subplots(3, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.set_title('Original pulse')
ax_orig.margins(0, 0.1)
ax_win.plot(win)
ax_win.set_title('Filter impulse response')
ax_win.margins(0, 0.1)
ax_filt.plot(filtered)
ax_filt.set_title('Filtered signal')
ax_filt.margins(0, 0.1)
fig.tight_layout()
fig.show()
```

### scipy.signal.spectrogram()

```python
import numpy as np
from scipy import signal
from scipy.fft import fftshift
import matplotlib.pyplot as plt
rng = np.random.default_rng()
```

```python
fs = 10e3
N = 1e5
amp = 2 * np.sqrt(2)
noise_power = 0.01 * fs / 2
time = np.arange(N) / float(fs)
mod = 500*np.cos(2*np.pi*0.25*time)
carrier = amp * np.sin(2*np.pi*3e3*time + mod)
noise = rng.normal(scale=np.sqrt(noise_power), size=time.shape)
noise *= np.exp(-time/5)
x = carrier + noise
```

```python
f, t, Sxx = signal.spectrogram(x, fs)
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

### scipy.interpolate.griddata()

```python
# Interpolando uma função 2D
import numpy as np
def func(x, y):
    return x*(1-x)*np.cos(4*np.pi*x) * np.sin(4*np.pi*y**2)**2
```

```python
# em uma grade de [0, 1]x[0, 1]
grid_x, grid_y = np.mgrid[0:1:100j, 0:1:200j]
```

```python
# mas só conhecemos seus valores em 1.000 pontos de dados
rng = np.random.default_rng()
points = rng.random((1000, 2))
values = func(points[:,0], points[:,1])
```

```python
# Isso pode ser feito com griddata testando todos os métodos de interpolação
from scipy.interpolate import griddata
grid_z0 = griddata(points, values, (grid_x, grid_y), method='nearest')
grid_z1 = griddata(points, values, (grid_x, grid_y), method='linear')
grid_z2 = griddata(points, values, (grid_x, grid_y), method='cubic')
```

```python
# Podemos ver que o resultado exato é reproduzido por todos os métodos até certo ponto, 
# mas para esta função suave o interpolante cúbico por partes dá os melhores resultados
import matplotlib.pyplot as plt
plt.subplot(221)
plt.imshow(func(grid_x, grid_y).T, extent=(0,1,0,1), origin='lower')
plt.plot(points[:,0], points[:,1], 'k.', ms=1)
plt.title('Original')
plt.subplot(222)
plt.imshow(grid_z0.T, extent=(0,1,0,1), origin='lower')
plt.title('Nearest')
plt.subplot(223)
plt.imshow(grid_z1.T, extent=(0,1,0,1), origin='lower')
plt.title('Linear')
plt.subplot(224)
plt.imshow(grid_z2.T, extent=(0,1,0,1), origin='lower')
plt.title('Cubic')
plt.gcf().set_size_inches(6, 6)
plt.show()
```

### scipy.special.factorial()

```python
import numpy as np
from scipy.special import factorial

arr = np.array([3, 4, 5])
factorial(arr, exact=False)

np.array([6., 24., 120.])
factorial(arr, exact=True)

np.array([6, 24, 120])
factorial(5, exact=True)
```

### scipy.special.comb()

```python
import numpy as np
from scipy.special import comb

k = np.array([3, 4])
n = np.array([10, 10])

comb(n, k, exact=False)
np.array([ 120.,  210.])

print(comb(10, 3, exact=True))

print(comb(10, 3, exact=True, repetition=True))
```

### scipy.special.gamma()

```python
import numpy as np
from scipy.special import gamma, factorial
import matplotlib.pyplot as plt
```

```python
gamma([0, 0.5, 1, 5])
```

```python
z = 2.5 + 1j
gamma(z)
```

```python
gamma(z+1), z*gamma(z)  # Propriedade de recorrência
```

```python
gamma(0.5)**2  # gamma(0.5) = sqrt(pi)
```

```python
# Plotando gama (x) para x real
x = np.linspace(-3.5, 5.5, 2251)
y = gamma(x)
```

```python
plt.plot(x, y, 'b', alpha=0.6, label='gamma(x)')
k = np.arange(1, 7)
plt.plot(k, factorial(k-1), 'k*', alpha=0.6,
         label='(x-1)!, x = 1, 2, ...')
plt.xlim(-3.5, 5.5)
plt.ylim(-10, 25)
plt.grid()
plt.xlabel('x')
plt.legend(loc='lower right')
plt.show()
```
